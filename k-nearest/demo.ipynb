## Классификация методом k-ближних
### Описание
Демо для демонстрации работы алгоритма классификации методом k-ближних на примере решения задачи ирисов Фишера.
### Алгоритм
Алгоритм k-ближайших соседей (kNN) является простым и интерпретируемым методом классификации и регрессии в машинном обучении. Он основывается на принципе "похожесть ближайших соседей": для классификации новой точки данных KNN находит k ближайших соседей из обучающего набора данных и присваивает ей класс, который наиболее часто встречается среди этих соседей (в случае классификации) или вычисляет среднее значение целевых переменных среди соседей (в случае регрессии). Алгоритм не требует обучения в привычном смысле и полагается только
на данные обучающего набора и выбранный параметр k. Выбор оптимального значения k может оказать влияние на качество модели: меньшие значения k могут привести к более шумным решениям, а большие значения k могут сделать модель более устойчивой, но менее точной.
![Algorythm visualisation](https://drive.google.com/uc?export=view&id=1aAegX-BUBHDnuCBcDj-q6GUINtrE290Y)

Плюсы алгоритма k-ближних:

+ **Простота реализации**: Алгоритм k-NN легко понимать и реализовывать. Это один из самых простых методов машинного обучения.

+ **Неявное обучение**: Нет необходимости в сложном обучении модели. Модель k-NN просто "запоминает" обучающие данные.

+ **Адаптивность к изменяющимся данным**: Модель может адаптироваться к изменениям в данных, поскольку она пересчитывает расстояния до ближайших соседей при каждом предсказании.

+ **Хорошо работает для небольших наборов данных**: В случае небольших наборов данных или данных с несложной структурой k-NN может давать хорошие результаты.

Минусы алгоритма k-ближних:

+ **Вычислительная сложность**: При большом объеме данных или большом числе признаков вычисление расстояний до всех точек может быть очень медленным.

+ **Чувствительность к выбору k**: Выбор значения k может сильно влиять на результаты. Слишком маленькое k может сделать модель чувствительной к шуму, а слишком большое - менее гибкой.

+ **Не учитывает важность признаков**: Метод не учитывает важность или вклад каждого признака в принятие решения, что может быть нежелательным для некоторых задач.

+ **Проблемы с балансом классов**: В случае несбалансированных классов, когда один класс представлен значительно больше другого, k-NN может быть предвзятым к более крупному классу.






```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.metrics import accuracy_score, silhouette_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Загрузка данных
iris = datasets.load_iris()
x = iris.data
labels = iris.target

# Разделение данных на обучающую и тестовую выборки
x_train, x_test, labels_train, labels_test = train_test_split(x, labels, test_size=0.3)
```

```python
# Создание KNN-классификатора с k=3
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, labels_train)
```

```python
# Предсказание классов на тестовой выборке
labels_pred = knn.predict(x_test)
```

```python
# Визуализация результатов кластеризации
plt.figure(figsize=(10, 5))

# График исходных данных
plt.subplot(1, 2, 1)
plt.scatter(x_train[:, 0], x_train[:, 1], c=labels_train)
plt.title("Trained")
plt.xlabel("petal lenght")
plt.ylabel("petal width")

# График с результатами кластеризации
plt.subplot(1, 2, 2)
plt.scatter(x_test[:, 0], x_test[:, 1], c=labels_test)
plt.title("Сlustering by the k-nearest method")
plt.xlabel("petal lenght")
plt.ylabel("petal width")

plt.tight_layout()
plt.show()
```

```python
# Оценка качества кластеризации
accuracy = accuracy_score(labels_test, labels_pred)
print(f"Accuracy:\t\t{accuracy:.2f}")
```
